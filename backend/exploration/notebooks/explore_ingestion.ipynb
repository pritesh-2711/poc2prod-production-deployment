```python

# We use **asyncpg** for async PostgreSQL access (consistent with FastAPI's async stack).

# Flow:
# 1. Connect → verify tables exist
# 2. Extract → hierarchical chunk → embed child chunks
# 3. INSERT parents into `parenthierarchy`, INSERT children into `ingestions`
# 4. Retrieve: embed query → cosine similarity search → fetch parent context
```

```python
import asyncpg
import os

DB_CONFIG = dict(
    host=os.getenv("DB_HOST", "localhost"),
    port=int(os.getenv("DB_PORT", 5432)),
    database=os.getenv("DB_NAME", "poc_to_prod"),
    user=os.getenv("DB_USER", "postgres"),
    password=os.getenv("DB_PASSWORD", ""),
)

async def get_conn():
    conn = await asyncpg.connect(
        **DB_CONFIG,
        server_settings={"search_path": "poc2prod,public"},
    )
    return conn

# Quick connectivity check
conn = await get_conn()
print("Connected:", conn.get_server_version())
await conn.close()
```

```python
# Verify that all expected tables exist in poc2prod schema
conn = await get_conn()

rows = await conn.fetch("""
    SELECT table_name
    FROM information_schema.tables
    WHERE table_schema = 'poc2prod'
    ORDER BY table_name;
""")

print("Tables in poc2prod schema:")
for r in rows:
    print(" ", r["table_name"])

await conn.close()
```

```python
# Inspect column definitions for each key table
conn = await get_conn()

for table in ("chats", "parenthierarchy", "ingestions"):
    cols = await conn.fetch("""
        SELECT column_name, data_type, is_nullable
        FROM information_schema.columns
        WHERE table_schema = 'poc2prod' AND table_name = $1
        ORDER BY ordinal_position;
    """, table)
    print(f"\n--- poc2prod.{table} ---")
    for c in cols:
        print(f"  {c['column_name']:<25} {c['data_type']:<20} nullable={c['is_nullable']}")

await conn.close()
```

#### End-to-end Ingestion

Steps:
1. Use already-extracted `text_records` and `HierarchicalChunker` (run earlier in notebook)
2. Get `(parent_docs, child_docs)` via `chunk_with_parents()`
3. Embed all child chunks with `OllamaEmbedder`
4. INSERT each unique parent into `parenthierarchy` → capture returned UUID
5. INSERT each child into `ingestions` with the matching `parent_id`

```python
type(text_records[0])
```

```python
import json as _json

# --- 1. Chunk hierarchically (parent + child docs) ---
chunker = HierarchicalChunker()
all_parent_docs = []
all_child_docs  = []

for record in text_records:
    if record.record_type not in ("text", "latex"):
        continue
    text = record.content
    if not isinstance(text, str) or not text.strip():
        continue
    meta = {"page": record.page, "type": record.record_type}
    if record.bbox:
        meta["bbox"] = record.bbox
    parents, children = chunker.chunk_with_parents(text, metadata=meta)
    all_parent_docs.extend(parents)
    all_child_docs.extend(children)

print(f"Parent chunks : {len(all_parent_docs)}")
print(f"Child chunks  : {len(all_child_docs)}")
```

```python
# --- 2. Embed all child chunks ---
embedder = OllamaEmbedder(model="nomic-embed-text-v2-moe:latest")

child_texts = [doc.page_content for doc in all_child_docs]
child_embeddings = embedder.embed(child_texts)   # list[list[float]]

print(f"Embedded {len(child_embeddings)} child chunks")
print(f"Embedding dimension: {len(child_embeddings[0])}")
```

```python
from src.core.config import ConfigManager
from src.embedding import LocalEmbedder, OllamaEmbedder, OpenAIEmbedder

cfg = ConfigManager(config_path='../configs/config.yaml')
emb_cfg = cfg.embedding_config

print(f'Active embedding provider : {emb_cfg.provider}')
print(f'Model                     : {emb_cfg.model}')
print(f'Vector dimension          : {emb_cfg.dimension}')

# Instantiate embedder that matches config.yaml — swap provider there, not here
_EMBEDDER_MAP = {
    'local':  lambda: LocalEmbedder(model=emb_cfg.model),
    'ollama': lambda: OllamaEmbedder(model=emb_cfg.model),
    'openai': lambda: OpenAIEmbedder(model=emb_cfg.model, api_key=emb_cfg.api_key),
}
embedder = _EMBEDDER_MAP[emb_cfg.provider]()
EMBEDDING_DIM = emb_cfg.dimension   # used for IVFFlat index creation below
```

```python
import uuid as _uuid

# --- 3. Insert parents into parenthierarchy ---
# asyncpg requires pgvector type to be passed as a string "[...]"
# We map each integer parent_id (from chunker) to the UUID returned by the DB.

filename = os.path.basename(pdf_path)

conn = await get_conn()
await conn.execute("SET search_path TO poc2prod, public;")

# parent_index_to_uuid: int -> UUID  (for linking children)
parent_index_to_uuid: dict[int, str] = {}

for idx, parent_doc in enumerate(all_parent_docs):
    meta = {k: v for k, v in parent_doc.metadata.items()}
    row = await conn.fetchrow(
        """
        INSERT INTO poc2prod.parenthierarchy (parent_chunk_content, filename, metadata)
        VALUES ($1, $2, $3)
        RETURNING id;
        """,
        parent_doc.page_content,
        filename,
        _json.dumps(meta),
    )
    parent_index_to_uuid[idx] = str(row["id"])

await conn.close()
print(f"Inserted {len(parent_index_to_uuid)} parent chunks into parenthierarchy")
```

```python
# --- 4. Insert child chunks into ingestions ---
# pgvector extension is registered via asyncpg codec — we pass embeddings as strings.

conn = await get_conn()
await conn.execute("SET search_path TO poc2prod, public;")

# Register pgvector type so asyncpg accepts LIST as vector
await conn.execute("SET LOCAL search_path TO poc2prod, public;")

inserted = 0
for child_doc, embedding in zip(all_child_docs, child_embeddings):
    meta      = child_doc.metadata
    parent_idx = meta.get("parent_id")   # integer from HierarchicalChunker
    parent_uuid = parent_index_to_uuid.get(parent_idx) if parent_idx is not None else None

    # Format embedding as pgvector string literal: '[0.1, 0.2, ...]'
    vec_str = "[" + ",".join(str(v) for v in embedding) + "]"

    child_meta = {k: v for k, v in meta.items() if k not in ("parent_id", "parent_text")}

    await conn.execute(
        """
        INSERT INTO poc2prod.ingestions
            (parent_id, filename, type, chunk_content, embeddings, metadata)
        VALUES
            ($1, $2, $3, $4, $5::vector, $6)
        """,
        _uuid.UUID(parent_uuid) if parent_uuid else None,
        filename,
        "pdf",
        child_doc.page_content,
        vec_str,
        _json.dumps(child_meta),
    )
    inserted += 1

await conn.close()
print(f"Inserted {inserted} child chunks into ingestions")
```

### Retrieval

Query flow:
1. Embed the user query
2. Cosine similarity search over `ingestions.embeddings` → top-k child chunks
3. For each matched child, fetch its parent chunk from `parenthierarchy`
4. Return parent text as context (richer window for the LLM)

```python
query = "What is the attention mechanism in transformers?"

# Embed the query (same embedder used during ingestion)
query_embedding = embedder.embed_one(query)
query_vec_str   = "[" + ",".join(str(v) for v in query_embedding) + "]"

TOP_K = 5

conn = await get_conn()

# Cosine similarity: operator <=> gives cosine distance (lower = more similar)
child_rows = await conn.fetch(
    """
    SELECT
        i.id          AS child_id,
        i.chunk_content,
        i.parent_id,
        1 - (i.embeddings <=> $1::vector) AS similarity
    FROM poc2prod.ingestions i
    WHERE i.embeddings IS NOT NULL
    ORDER BY i.embeddings <=> $1::vector
    LIMIT $2;
    """,
    query_vec_str,
    TOP_K,
)

print(f"Top-{TOP_K} matched child chunks:\n")
for r in child_rows:
    print(f"  similarity={r['similarity']:.4f}  child_id={r['child_id']}")
    print(f"  chunk: {r['chunk_content'][:120]}...")
    print()

await conn.close()
```

```python
# Fetch parent context for each matched child
parent_ids = list({str(r["parent_id"]) for r in child_rows if r["parent_id"]})

conn = await get_conn()

parent_rows = await conn.fetch(
    """
    SELECT id, parent_chunk_content, filename
    FROM poc2prod.parenthierarchy
    WHERE id = ANY($1::uuid[]);
    """,
    parent_ids,
)

parent_map = {str(r["id"]): r for r in parent_rows}

print("Parent context retrieved for matched chunks:\n")
for r in child_rows:
    pid = str(r["parent_id"]) if r["parent_id"] else None
    parent = parent_map.get(pid)
    print(f"  child_id  : {r['child_id']}")
    print(f"  similarity: {r['similarity']:.4f}")
    print(f"  child text: {r['chunk_content'][:100]}...")
    if parent:
        print(f"  parent text (first 200 chars): {parent['parent_chunk_content'][:200]}...")
    print()

await conn.close()
```
