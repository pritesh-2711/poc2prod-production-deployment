````python
# PDFPlumber - Convert Tables to Structured Formats
# Once extracted, tables can be stored in multiple formats for RAG:

def extract_and_structure_tables(pdf_path):
    """
    Extract tables from PDF and convert to structured formats
    Handles duplicate column names by making them unique.
    Returns: dict with page_number, table_index, dataframe, csv_string, json_string
    """
    from collections import Counter
    
    structured_tables = []
    
    def make_unique_columns(columns):
        """Handle duplicate column names by appending suffixes"""
        counts = Counter(columns)
        result = []
        column_counters = Counter()
        
        for col in columns:
            if counts[col] > 1:
                column_counters[col] += 1
                result.append(f"{col}_{column_counters[col]}")
            else:
                result.append(col)
        return result
    
    with pdfplumber.open(pdf_path) as pdf:
        for page_num, page in enumerate(pdf.pages):
            tables = page.extract_tables()
            
            if not tables:
                continue
            
            for table_idx, table in enumerate(tables):
                if len(table) < 2:  # Skip malformed tables
                    continue
                
                # Create DataFrame with original columns
                df = pd.DataFrame(table[1:], columns=table[0])
                original_columns = list(df.columns)
                
                # Make columns unique if needed
                unique_columns = make_unique_columns(original_columns)
                df.columns = unique_columns
                
                try:
                    # Store in multiple formats
                    structured_tables.append({
                        'page': page_num + 1,
                        'table_index': table_idx,
                        'dataframe': df,
                        'csv': df.to_csv(index=False),
                        'json': df.to_json(orient='records'),
                        'markdown': df.to_markdown(index=False),
                        'columns': unique_columns,
                        'original_columns': original_columns,
                        'rows': len(df),
                        'metadata': {
                            'source': pdf_path.split('/')[-1],
                            'page_number': page_num + 1,
                            'table_id': f"page_{page_num + 1}_table_{table_idx}",
                            'has_duplicate_columns': len(original_columns) != len(set(original_columns))
                        }
                    })
                except Exception as e:
                    print(f"Error processing table on page {page_num + 1}, table {table_idx}: {str(e)}")
                    continue
    
    return structured_tables
````

````python
# Detailed Image Extraction with PyMuPDF
# PyMuPDF (fitz) gives access to image XObjects and their metadata

pdf_file_path = data_directory + filename
doc = fitz.open(pdf_file_path)

print(f"Total pages: {len(doc)}\n")

image_metadata = []

for page_num in range(len(doc)):
    page = doc[page_num]
    
    # Get all image XObjects on the page
    image_list = page.get_images()
    
    if image_list:
        print(f"Page {page_num + 1}: Found {len(image_list)} image(s)")
        
        for img_index, img_ref in enumerate(image_list):
            # img_ref is a tuple: (xref_number,)
            xref = img_ref[0]
            
            # Get image information
            pix = fitz.Pixmap(doc, xref)
            
            # Extract basic metadata
            img_info = {
                'page': page_num + 1,
                'image_index': img_index,
                'xref': xref,
                'width': pix.width,
                'height': pix.height,
                'size_pixels': pix.width * pix.height,
                'n': pix.n,  # Number of components (3=RGB, 4=RGBA, 1=Gray, etc.)
                'colorspace': pix.colorspace.name if pix.colorspace else None,
                'alpha': pix.alpha,  # Has alpha channel
            }
            
            # Get image bytes and calculate compressed size
            img_bytes = pix.tobytes()
            img_info['uncompressed_bytes'] = len(img_bytes)
            
            # Calculate image hash for deduplication
            img_info['hash'] = hashlib.md5(img_bytes).hexdigest()
            
            # Get aspect ratio
            img_info['aspect_ratio'] = pix.width / pix.height if pix.height > 0 else 0
            
            print(f"  Image {img_index + 1}:")
            print(f"    Dimensions: {img_info['width']}x{img_info['height']} px")
            print(f"    Colorspace: {img_info['colorspace']}")
            print(f"    Size: {img_info['uncompressed_bytes'] / 1024:.1f} KB")
            print(f"    Aspect ratio: {img_info['aspect_ratio']:.2f}")
            
            image_metadata.append(img_info)
            
            # Clean up
            pix = None

doc.close()
print(f"\nTotal images extracted: {len(image_metadata)}")
````

````python
import json
import re
from pathlib import Path
from typing import Any
````

````python
from docling.document_converter import DocumentConverter, PdfFormatOption
from docling.datamodel.base_models import InputFormat
from docling.datamodel.pipeline_options import PdfPipelineOptions
````

````python
URL_PATTERN = re.compile(
    r"""(?ix)\b(
        (?:https?://|www\.)
        [^\s<>()\[\]{}"']+
    )"""
)
````

````python
def safe_get_page(item: Any) -> int | None:
    prov = getattr(item, "prov", None)
    if prov and len(prov) > 0:
        return getattr(prov[0], "page_no", None)
    return None
````

````python
def safe_get_bbox(item: Any) -> dict | None:
    prov = getattr(item, "prov", None)
    if prov and len(prov) > 0 and getattr(prov[0], "bbox", None):
        bbox = prov[0].bbox
        return {
            "l": getattr(bbox, "l", None),
            "t": getattr(bbox, "t", None),
            "r": getattr(bbox, "r", None),
            "b": getattr(bbox, "b", None),
        }
    return None
````

````python
def extract_urls_from_text(text: str) -> list[str]:
    if not text:
        return []
    return list({m.group(1) for m in URL_PATTERN.finditer(text)})
````

````python
def normalize_docling_item(item: Any) -> list[dict]:
    """
    Returns one or more normalized records.
    A text item may also emit url records if URLs are found inside it.
    """
    out: list[dict] = []

    item_type = item.__class__.__name__
    page = safe_get_page(item)
    bbox = safe_get_bbox(item)

    # TEXT / FORMULA
    if item_type == "TextItem":
        label = str(getattr(item, "label", "")).upper()
        text = getattr(item, "text", "") or ""

        if "FORMULA" in label:
            # Formula enrichment may populate extra fields depending on version/model.
            # Keep raw text, and if enriched latex exists, capture it.
            latex = getattr(item, "latex", None) or getattr(item, "orig", None) or text
            out.append({
                "type": "latex",
                "page": page,
                "bbox": bbox,
                "content": latex,
                "raw": text,
            })
        else:
            out.append({
                "type": "text",
                "page": page,
                "bbox": bbox,
                "content": text,
                "raw": text,
            })

            for url in extract_urls_from_text(text):
                out.append({
                    "type": "url",
                    "page": page,
                    "bbox": bbox,
                    "content": url,
                    "raw": text,
                })

    # TABLE
    elif item_type == "TableItem":
        table_data = None
        try:
            # Often available via export methods or nested data
            if hasattr(item, "export_to_markdown"):
                table_data = item.export_to_markdown()
            elif hasattr(item, "data"):
                table_data = str(item.data)
        except Exception:
            table_data = None

        out.append({
            "type": "table",
            "page": page,
            "bbox": bbox,
            "content": table_data,
            "raw": item_type,
        })

    # IMAGE / PICTURE
    elif item_type == "PictureItem":
        out.append({
            "type": "image",
            "page": page,
            "bbox": bbox,
            "content": {
                "caption": getattr(item, "caption_text", None),
                "annotations": getattr(item, "annotations", None),
            },
            "raw": item_type,
        })

    return out
````

````python
def run_docling(pdf_path: str) -> dict:
    pipeline_options = PdfPipelineOptions()

    # Turn on useful enrichments
    pipeline_options.do_formula_enrichment = False
    pipeline_options.do_picture_classification = True
    # optionally:
    # pipeline_options.do_picture_description = True

    converter = DocumentConverter(
        format_options={
            InputFormat.PDF: PdfFormatOption(pipeline_options=pipeline_options)
        }
    )

    result = converter.convert(pdf_path)
    doc = result.document

    records: list[dict] = []

    # DoclingDocument contains collections for texts/tables/pictures depending on version,
    # but walking exported dict is often the most version-tolerant method.
    doc_dict = doc.export_to_dict()

    # Version-tolerant traversal
    for key in ("texts", "tables", "pictures"):
        for item in doc_dict.get(key, []):
            label = str(item.get("label", "")).upper()
            page = None
            bbox = None

            prov = item.get("prov", [])
            if prov:
                page = prov[0].get("page_no")
                bbox = prov[0].get("bbox")

            if key == "texts":
                text = item.get("text", "")
                if "FORMULA" in label:
                    records.append({
                        "type": "latex",
                        "page": page,
                        "bbox": bbox,
                        "content": item.get("latex") or text,
                        "raw": text,
                    })
                else:
                    records.append({
                        "type": "text",
                        "page": page,
                        "bbox": bbox,
                        "content": text,
                        "raw": text,
                    })
                    for url in extract_urls_from_text(text):
                        records.append({
                            "type": "url",
                            "page": page,
                            "bbox": bbox,
                            "content": url,
                            "raw": text,
                        })

            elif key == "tables":
                records.append({
                    "type": "table",
                    "page": page,
                    "bbox": bbox,
                    "content": item,
                    "raw": "table",
                })

            elif key == "pictures":
                records.append({
                    "type": "image",
                    "page": page,
                    "bbox": bbox,
                    "content": item,
                    "raw": "picture",
                })

    return {
        "source": str(pdf_path),
        "records": records,
    }
````

````python
def reconstruct_table(table_content: dict) -> pd.DataFrame:
    """
    Reconstruct a pandas DataFrame from Docling's extracted table content.

    Each cell in `data.table_cells` carries its grid position via
    start_row_offset_idx / start_col_offset_idx, so we can place every
    cell into the correct slot of a 2-D grid and then build a DataFrame.
    """
    cells = table_content["data"]["table_cells"]

    # Determine grid dimensions
    num_rows = max(c["end_row_offset_idx"] for c in cells)
    num_cols = max(c["end_col_offset_idx"] for c in cells)

    # Build an empty grid
    grid = [[""] * num_cols for _ in range(num_rows)]

    # Track which rows are purely column-header rows
    header_rows = set()

    for cell in cells:
        r = cell["start_row_offset_idx"]
        c = cell["start_col_offset_idx"]
        grid[r][c] = cell["text"]
        if cell.get("column_header"):
            header_rows.add(r)

    if header_rows:
        # Use the last header row as column names
        header_row_idx = max(header_rows)
        columns = grid[header_row_idx]
        data_rows = [grid[i] for i in range(num_rows) if i not in header_rows]
    else:
        columns = [f"col_{i}" for i in range(num_cols)]
        data_rows = grid

    return pd.DataFrame(data_rows, columns=columns)
````

````python
# ── Types that are worth summarising ──────────────────────────────────────────
PROCESS_TYPES = {
    "bar_chart", "pie_chart", "line_chart", "scatter_plot", "box_plot",
    "photograph", "full_page_image", "flow_chart",
}
MIN_CONFIDENCE = 0.40   # ignore the top class if the model is unsure
````

````python
# ── helper: extract bbox from image content dict ──────────────────────────────
def get_bbox_from_content(content: dict) -> dict | None:
    prov = content.get("prov", [])
    if prov:
        return prov[0].get("bbox")
    return None
````

````python
# ── 1. Decide whether this image is worth a vision call ───────────────────────
def should_process_image(image_content: dict) -> tuple[bool, str]:
    """
    Returns (True, top_class) if the image should be summarised,
    (False, top_class) otherwise.
    Uses the highest-confidence prediction from Docling's classifier.
    """
    predictions = (
        image_content
        .get("meta", {})
        .get("classification", {})
        .get("predictions", [])
    )
    if not predictions:
        return False, "unknown"

    top = max(predictions, key=lambda p: p["confidence"])
    top_class = top["class_name"]
    confidence = top["confidence"]

    if confidence < MIN_CONFIDENCE:
        return False, top_class
    return top_class in PROCESS_TYPES, top_class
````

````python
# ── 2. Crop the image out of the PDF page ─────────────────────────────────────
def crop_image_from_pdf(pdf_path: str, page_no: int, bbox: dict) -> bytes:
    """
    Extracts and returns the image region as PNG bytes.

    Docling stores bbox with coord_origin='BOTTOMLEFT', so we convert:
        y_top_from_top    = page_height - bbox['t']
        y_bottom_from_top = page_height - bbox['b']
    """
    doc = fitz.open(pdf_path)
    page = doc[page_no - 1]           # Docling page_no is 1-indexed
    page_height = page.rect.height

    rect = fitz.Rect(
        bbox["l"],
        page_height - bbox["t"],      # top of image in PyMuPDF coords
        bbox["r"],
        page_height - bbox["b"],      # bottom of image in PyMuPDF coords
    )

    # Render at 2× resolution for better OCR/vision quality
    mat = fitz.Matrix(2, 2)
    clip = page.get_pixmap(matrix=mat, clip=rect)
    doc.close()
    return clip.tobytes("png")
````

````python
# # ── 3. Pull surrounding text from the same page ───────────────────────────────

def get_image_context(df, page_no: int, bbox: dict, n_chars: int = 600) -> str:
    img_mid = (bbox["t"] + bbox["b"]) / 2

    page_text = df[(df["type"] == "text") & (df["page"] == page_no)].copy()
    if page_text.empty:
        return ""

    def row_mid(row):
        # bbox is now inside content for text records — but text records
        # store content as a plain string, so we skip bbox-based sorting
        # and just return all text on the page ordered by df position
        return 0

    # For text records bbox isn't in df either; return all page text as context
    return df[(df["type"] == "text") & (df["page"] == page_no)]["content"].str.cat(sep=" ")[:n_chars]
````

````python
from openai import OpenAI

def summarize_image(
    image_bytes: bytes,
    image_type: str,
    context: str,
    client: OpenAI,
) -> dict:
    b64 = base64.standard_b64encode(image_bytes).decode()

    prompt = f"""You are analysing an image extracted from a corporate annual report PDF.

Image type (from classifier): {image_type}

Surrounding text from the same page:
{context}

Answer the following in JSON with exactly these keys:
{{
  "description": "What is visually shown in this image?",
  "document_intent": "What is the document trying to communicate through this image?",
  "context_link": "Is this image self-contained or does it directly support the surrounding text? How?",
  "key_data_points": ["list any numbers, labels, or named entities visible in the image"]
}}

Return only the JSON object, no extra text."""

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {
                "role": "user",
                "content": [
                    {
                        "type": "image_url",
                        "image_url": {
                            "url": f"data:image/png;base64,{b64}",
                            "detail": "high",
                        },
                    },
                    {"type": "text", "text": prompt},
                ],
            }
        ],
        max_tokens=1024,
    )

    raw = response.choices[0].message.content.strip()
    # strip markdown code fences if model wraps the JSON
    if raw.startswith("```"):
        raw = raw.split("```")[1]
        if raw.startswith("json"):
            raw = raw[4:]
    try:
        return json.loads(raw)
    except json.JSONDecodeError:
        return {"raw_response": raw}
````

````python
# ── 5. Orchestrator — runs across all images in df ────────────────────────────
def process_images(df, pdf_path: str, client: OpenAI) -> list[dict]:
    results = []

    image_rows = df[df["type"] == "image"]
    print(f"Total images found: {len(image_rows)}")

    for _, row in image_rows.iterrows():
        content = row["content"]
        page_no = row["page"]

        # bbox comes from inside the content dict, not from df
        bbox = get_bbox_from_content(content)
        if bbox is None:
            print(f"  Skipping page {page_no} — no bbox found")
            continue

        should_run, top_class = should_process_image(content)

        if not should_run:
            print(f"  Skipping page {page_no} — type: {top_class}")
            continue

        print(f"  Processing page {page_no} — type: {top_class}")

        image_bytes = crop_image_from_pdf(pdf_path, page_no, bbox)
        context = get_image_context(df, page_no, bbox)
        summary = summarize_image(image_bytes, top_class, context, client)

        results.append({
            "self_ref": content.get("self_ref"),
            "page": page_no,
            "type": top_class,
            **summary,
        })

    return results
````
